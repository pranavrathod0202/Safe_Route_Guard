```python
import matplotlib.pyplot as plt
# Create a new column 'Accident Group' based on the number of accidents
df['Accident Group'] = pd.cut(df['No. of Accident'], bins=[0, 5, 10, 100], labels=['1-5', '6-10', '>10'])

# Group the DataFrame by the 'Accident Group' column and count the number of locations for each group
grouped_df = df.groupby('Accident Group')['Location'].nunique()

# Print the grouped DataFrame
print(grouped_df)
# Group the DataFrame by the 'Accident Group' column and count the number of locations for each group
grouped_df = df.groupby('Accident Group')['Location'].nunique().reset_index(name='counts')

# Plot the bar chart
plt.bar(grouped_df['Accident Group'], grouped_df['counts'])
plt.xlabel('Accident Group')
plt.ylabel('Number of Locations')
plt.title('Number of Locations by Accident Group')
plt.show()
```

```python
import matplotlib.pyplot as plt

df['Fatalities Group'] = pd.cut(df['No. of Fatalities'], bins=[0, 5, 10, 100], labels=['1-5', '6-10', '>10'])


grouped_df = df.groupby('Fatalities Group')['Location'].nunique()

print(grouped_df)

grouped_df = df.groupby('Fatalities Group')['Location'].nunique().reset_index(name='counts')


plt.bar(grouped_df['Fatalities Group'], grouped_df['counts'])
plt.xlabel('Fatalities')
plt.ylabel('Number of Locations')
plt.title('Number of Locations by Fatalities')
```

```python
grouped_df = df.groupby('No. of Accident')['No. of Fatalities'].mean().reset_index()

# Plot the line graph
plt.plot(grouped_df['No. of Accident'], grouped_df['No. of Fatalities'])
plt.xlabel('Number of Accidents')
plt.ylabel('Number of Fatalities')
plt.title('Number of Fatalities by Number of Accidents')
plt.grid(True)
plt.show()
```
