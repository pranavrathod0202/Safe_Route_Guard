```python
import matplotlib.pyplot as plt
# Create a new column 'Accident Group' based on the number of accidents
df['No.of Vehicles Group'] = pd.cut(df['Number_of_Vehicles'], bins=[0, 5, 10, 100], labels=['1-2', '2-3', '>3'])

# Group the DataFrame by the 'Accident Group' column and count the number of locations for each group
grouped_df = df.groupby('No.of Vehicles Group')['Number_of_Casualties'].nunique()

# Print the grouped DataFrame
print(grouped_df)
# Group the DataFrame by the 'Accident Group' column and count the number of locations for each group
grouped_df = df.groupby('Number_of_Vehicles')['Number_of_Casualties'].nunique().reset_index(name='counts')

# Plot the bar chart
plt.bar(grouped_df['Number_of_Vehicles'], grouped_df['counts'])
plt.xlabel('Number_of_Vehicles')
plt.ylabel('Number_of_Casualtiess')
plt.title('Number of Vehicles and Casualty')
plt.show()
```

```python
import matplotlib.pyplot as plt


grouped_df = df.groupby('Number_of_Vehicles')['Number_of_Casualties'].mean().reset_index()

# Plot the line graph
plt.plot(grouped_df['Number_of_Vehicles'], grouped_df['Number_of_Casualties'])
plt.xlabel('Number of Accidents')
plt.ylabel('Number of Fatalities')
plt.title('Number of Fatalities by Number of Accidents')
plt.grid(True)
plt.show()
```
